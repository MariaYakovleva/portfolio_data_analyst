## Загрузка и подготовка исходных данных

### Описание данных

В .csv файле `gym_churn` представлена следующая информация:

**Данные клиента за предыдущий до проверки факта оттока месяц:**
- 'gender' — пол;
- 'Near_Location' — проживание или работа в районе, где находится фитнес-центр;
- 'Partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
- Promo_friends — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
- 'Phone' — наличие контактного телефона;
- 'Age' — возраст;
- 'Lifetime' — время с момента первого обращения в фитнес-центр (в месяцах).<br>

**Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:**
- 'Contract_period' — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- 'Month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах);
- 'Group_visits' — факт посещения групповых занятий;
- 'Avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента;
- 'Avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц;
- 'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
- 'Churn' — факт оттока в текущем месяце.

### Построение матрицы корреляций

```python
cm = gym_churn.corr() #вычисляем матрицу корреляций



plt.figure(figsize=(10, 10))



#нарисуем тепловую карту с подписями для матрицы корреляций
sns.heatmap(cm, 
             annot=True, 
             square=True, 
             fmt='.2g', 
            cmap= 'coolwarm', 
             linewidths=3, 
            linecolor='black',
             annot_kws={'size':10})

plt.title('Матрица корреляций', fontsize = 22 )
plt.show()
```

* Признак *длительность текущего действующего абонемента* зависит от признака *срок до окончания текущего действующего абонемента* и наоборот
* Признак *средняя частота посещений в неделю за предыдущий месяц* зависит от признака *средняя частота посещений в неделю за все время с начала действия абонемента* <br>

Важным условием логистической регрессии является независимость переменных друг от друга. Отберем вручную признаки, чтобы избавиться от **мультиколлинеарности**.

По матрице корреляций понятно, что с оттоком клиентов коррелириют следующие показатели:
- Слабая отрицательная корреляция между фактом оттока и  временем с момента первого обращения в фитнес-центр: -0,44
- Слабая отрицательная корреляция между фактом оттока и длительностью текущего действующего абонемента: -0,39
- Слабая отрицательная корреляция между фактом оттока и средней частотой посещений в неделю за предыдущий месяц: -0,41 <br>
- Слабая отрицательная корреляция между фактом оттока и возрастом клиента:-0,4
- Слабая отрицательная корреляция между фактом оттока и сроком до окончания текущего действующего абонемента: -0,38 <br>

    <li>  средняя частота посещений в неделю за предыдущий месяц. Потерянные клиенты в большинстве своем вообще не посещали фитнес-зал </li>



## Построение модели прогнозирования оттока клиентов

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце

```python
#разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = gym_churn.drop(columns=['Churn'], axis=1)
y = gym_churn['Churn']

#добавим матрицу X_without_multu, без мультиколлинеарных признаков
X_without_multu = gym_churn.drop(columns=['Churn', 'Contract_period', 'Avg_class_frequency_current_month'], axis=1) 

#разделяем модель на обучающую и валидационную выборки в отношении 80/20
#зададим параметр stratify=y, для обеспечения и в учебных, и в проверочных наборах 
# данных одно и то же соотношение 0 и 1, что в исходном наборе данных.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify= y)

```

Обучим модель на train-выборке двумя способами:
- логистической регрессией,
- случайным лесом.

Стандартизируем данные.

```python
# обучим StandartScaler на обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)

# Преобразуем обучающий и валидационные наборы данных к стандартизированным
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

### Логистическая регрессия

```python
#зададим алгоритм LogisticRegression для нашей модели
model = LogisticRegression(solver='liblinear') 

#обучим модель
model.fit(X_train_st, y_train) 

# получим прогноз класса после обучения
predictions = model.predict(X_test_st ) 

#посчитаем вероятность принадлежности объекта классу 1
probabilities = model.predict_proba(X_test_st)[:,1] 


# выведем необходимые метрики 
print('Метрики для логистической регрессии')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
```

Доля правильных ответов близка к 100% <br>
90% прогнозов относительно класса 1 верна <br>
85% реальных объектов класса 1 мы смогли обнаружить.

### Случайный лес

```python
# зададим алгоритм для новой модели - случайный лес
clf_model = RandomForestClassifier(random_state=0)

#обучим модель
clf_model.fit(X_train_st, y_train) 

# воспользуемся уже обученной моделью, чтобы сделать прогнозы
clf_predictions =  clf_model.predict(X_test_st)
clf_probabilities = clf_model.predict_proba(X_test_st)[:,1] 

# выведем все изученные метрики
print('Метрики для дерева принятия решения')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, clf_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, clf_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, clf_predictions)))
```

Доля правильных ответов 92% <br>
88% прогнозов относительно класса 1 верна <br>
83% реальных объектов класса 1 мы смогли обнаружить.

Алгоритм обучения "Случайный лес" по показателям метрик accuracy, precision, recall показал себя хуже.
